## 1장 : 머신러닝과 딥러닝

### 1.1 : 인공지능, 머신러닝과 딥러닝
인공지능 >> 머신러닝 >> 딥러닝 순서
- 머신 러닝은 데이터를 인간이 전처리하는 과정이 필요함
- 딥러닝은 데이터를 신경망에 적용하면 컴퓨터가 분석 후 답을 찾음

### 1.2 : 머신러닝이란
학습 단계와 예측 단계로 구분함
학습 단계 = 모델링, 모델 평가, 모델 업데이트 & 검증 -- 최선의 모델 찾음
예측 단계 = 테스트 데이터셋으로 적용

* 머신러닝 알고리즘 = 지도 / 비지도 / 강화 학습
  1. 강화 학습 : 특정 환경에서 자신의 행동에 대한 보상을 받으며 학습 진행.

### 1.3 : 딥러닝이란
= 인간의 신경망 원리를 모방해 고안된 머신러닝 방법의 일종
- 데이터 준비 >> 모델 정의 > 모델 컴파일 > 모델 훈련 > 모델 예측
  * 은닉층이 많을 수록 성능은 좋아지지만 과적합 발생 확률 증가 -- 적절한 갯수를 선택해야 한다.

  * 모델 컴파일 과정 = 활성화 함수, 손실 함수, 옵티마이저 선택  

  * 모델 훈련 = 한 번에 처리할 데이터양을 결정 (batch와 epoch)

- 심층 신경망와 역전파. 역전파 = 가중치 업데이트 시 더 쉽게 하기 위한 것

- 딥러닝 학습 알고리즘
  - 지도 학습 : {CNN(합성곱 신경망), AlexNet, ResNet}, {RNN(순환 신경망), LSTM}
  - 비지도 학습 : {워드 임베딩(Word2Vec, Glove)}, {군집(가우시안 혼합 모델 - GMM, 자기 조직화 지도 - SOM)}, {차원축소 - 오토인코더, PCA}
  - 전이 학습 : 이미 학습 돼 있는 모델  - {BERT, MobileNetV2}
  - 강화 학습 : 마르코프 결정 과정

### 2.1 : 파이토치 개요
파이토치 = GPU에서 텐서 조작 및 동적 신경망 구축이 가능한 프레임워크

GPU - CUDA, cuDNN API로 GPU를 연산에 적용 가능
텐서 = 파이토치의 데이터 형태, 3차원 이상의 배열 형태 (행렬의 다차원 표현)

#### 2.1.2 : 파이토치 아키텍처
: 사용자 라이브러리 = 파이토치 API >> 파이토치 엔진 (AutoGrad - C++, Aten - C++, JIT - C++) >> 연산 처리 (텐서에 대한 연산 처리)

1. 사용자 라이브러리
  * torch
  * torch.autograd : 자동 미분 패키지
  * torch,nn : 신경망 구축 및 훈련 패키지
  * torch.multiprocessing : 파이썬 멀티프로세싱 패키지
  * torch.utils : DataLoader 및 기타 유틸리티 제공

2. 파이토치 엔진
  * Autograd C++ : 가중치와 바이어스 업데이트 과정에서 필요한 미분을 자동으로 계산
  * Aten C++ : C++ 텐서 라이브러리 제공
  * JIT C++ : 계산 최적화 관련 컴파일러

3. 연산처리 : CPU, GPU, THC를 이용해 연산 처리


#### 텐서

4. 모델 정의 :
  * 보통 nn.Module을 상속받은 클래스로 모델을 정의함.
  * 단순한 경우에는 nn.Module 없이 직접 함수형으로 모델을 정의할 수도 있음.
  * Sequential을 사용하면 층을 순차적으로 쌓는 신경망을 간단히 만들 수 있고,
  * 함수형 정의 방식은 변수에 저장된 계층을 재사용할 수 있는 장점이 있음.

5. 학습 구성 요소 :
  * 모델 파라미터: 학습을 통해 업데이트되는 w, b 등의 값.
  * 손실 함수 (Loss function): 모델의 출력과 실제 값의 차이를 측정함.
  * 옵티마이저 (Optimizer): 손실을 최소화하도록 파라미터를 업데이트함
  * 학습률 스케줄러 (Learning rate scheduler): 정해진 에포크마다 학습률을 조정함.
  * 지표 (Metric): 훈련 및 테스트 성능을 모니터링하기 위한 평가 기준임.

6. 모델 훈련 :
  * 기본적으로는 y = wx + b에서 적절한 w와 b를 찾는 과정임.
  * 기울기(gradient)를 초기화 → 자동 계산 → 업데이트하는 순서로 진행됨.

7. 모델 평가 및 모니터링 :

```python
import torch.nn as nn
class MLP(nn.Module):
    def __init__(self):
        super(MLP, self).__init__()
        self.layer1=nn.Sequential(
            nn.Conv2d(in_channels=3, out_channels=64, kernel_size=5),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2))

        self.layer2=nn.Sequential(
            nn.Conv2d(in_channels=64, out_channels=30, kernel_size=5),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2))

        self.layer3=nn.Sequential(
            nn.Linear(in_features=30*5*5, out_features=10, bias=True),
            nn.ReLU(inplace=True))

    def forward(self,x):
        x=self.layer1(x)
        x=self.layer2(x)
        x=x.view(x.shape[0],-1)
        x=self.layer3(x)
        return x
model=MLP()

print("Printing children\n-----------------------")
print(list(model.children()))
print("\n\nPrinting Modules\n-------------------------")
print(list(model.modules()))
```

```python
def MLP(in_features=1, hidden_features=20, out_features=1):
    hidden=nn.Linear(in_features=in_features, out_features=hidden_features,
                     bias=True)
    activation=nn.ReLU()
    output=nn.Linear(in_features=hidden_features, out_features=out_features,
                     bias=True)
    net=nn.Sequential(hidden, activation, output)
    return net
```
